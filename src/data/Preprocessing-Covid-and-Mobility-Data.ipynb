# Finding missing values for mobility data

```python
print(mobility_df.isnull().sum())
```

```python
#Dropping columns with more than 80% missing data
mobility_df = mobility_df.drop(columns = ['metro_area', 'iso_3166_2_code', 'census_fips_code'])
```

```python
#We don't need sub regions as well
mobility_df = mobility_df.drop(columns = ['sub_region_1', 'sub_region_2'])
```

```python
#Removing some null value rows
mobility_df.dropna(subset = ['country_region_code'], inplace = True)
mobility_df.dropna(subset = ['place_id'], inplace = True)
```

# Missing values in Covid data

```python
print(covid_df.isnull().sum())
```

```python
print(covid_df.shape)
```

```python
#Dropping columns with more than 75% missing values
threshold = 0.75 * len(covid_df)
covid_df.dropna(thresh=threshold, axis=1, inplace=True)
```

```python
print(covid_df.shape)
```

```python
covid_df.head()
```

```python
print(covid_df.isnull().sum())
```

```python
#For continent there's no need of imputation
covid_df.dropna(subset=['continent'], inplace=True)
```

```python
#For numerical columns, we will impute the missing values with the mean value
numerical_columns = ['total_cases', 'new_cases', 'total_deaths', 'new_deaths', 'total_cases_per_million', 
                     'new_cases_per_million', 'total_deaths_per_million', 'new_deaths_per_million', 
                     'population_density', 'median_age', 'aged_65_older', 'aged_70_older', 'gdp_per_capita',
                     'cardiovasc_death_rate', 'diabetes_prevalence', 'life_expectancy']

for column in numerical_columns:
    covid_df[column].fillna(covid_df[column].mean(), inplace=True)
```

```python
covid_df['human_development_index'].fillna(covid_df['human_development_index'].mean(), inplace=True)
```

```python
print(covid_df.isnull().sum())
```

```python
print(covid_df.shape)
```

```python
covid_df.head()
```

```python
#Doing ffill for remaining missing values, since the percentage is less
columns_to_ffill = ['new_cases_smoothed', 'new_deaths_smoothed', 
                    'new_cases_smoothed_per_million', 'new_deaths_smoothed_per_million']

for column in columns_to_ffill:
    covid_df[column].fillna(method='ffill', inplace=True)
```

```python
print(covid_df.isnull().sum())
```

```python
#Drop small percentage of missing rows
columns_with_missing = ['new_cases_smoothed', 'new_deaths_smoothed', 'new_deaths_smoothed_per_million']

covid_df.dropna(subset=columns_with_missing, inplace=True)
```

# Creating a new CSV, aggregating based on date

```python
covid_df.head()
```

```python
# Change the day of the date to '01' for both DataFrames
covid_df['date'] = covid_df['date'].apply(lambda x: x.replace(day=1))
mobility_df['date'] = mobility_df['date'].apply(lambda x: x.replace(day=1))
```

```python
mobility_df.head()
```

# Aggregating columns based on date

```python
covid_aggregation_columns = {
    'iso_code': 'first',
    'continent': 'first',
    'total_cases': 'sum',
    'new_cases': 'sum',
    'total_deaths': 'sum',
    'new_deaths': 'sum',
    'total_cases_per_million': 'mean',
    'new_cases_per_million': 'mean',
    'total_deaths_per_million': 'mean',
    'new_deaths_per_million': 'mean',
    'population_density': 'mean',
    'median_age': 'mean',
    'aged_65_older': 'mean',
    'aged_70_older': 'mean',
    'gdp_per_capita': 'mean',
    'cardiovasc_death_rate': 'mean',
    'diabetes_prevalence': 'mean',
    'life_expectancy': 'mean',
    'human_development_index': 'mean',
    'population': 'mean'
}


mobility_aggregation_columns = {
    'country_region_code': 'first',
    'retail_and_recreation_percent_change_from_baseline': 'mean',
    'grocery_and_pharmacy_percent_change_from_baseline': 'mean',
    'parks_percent_change_from_baseline': 'mean',
    'transit_stations_percent_change_from_baseline': 'mean',
    'workplaces_percent_change_from_baseline': 'mean',
    'residential_percent_change_from_baseline': 'mean'
}
```

```python
aggregated_cov_df = covid_df.groupby(['location', 'date']).agg(covid_aggregation_columns).reset_index()
aggregated_mobi_df = mobility_df.groupby(['country_region', 'date']).agg(mobility_aggregation_columns).reset_index()

```

```python
print(len(aggregated_cov_df))
print(len(aggregated_mobi_df))
```

# Merge the two dataframes now based on aggregated Loc and Date

```python
merged_date_df = aggregated_cov_df.merge(aggregated_mobi_df, left_on=['location', 'date'], right_on=['country_region', 'date'], how='right')
mobility_columns = ['country_region', 'country_region_code', 'retail_and_recreation_percent_change_from_baseline', 'grocery_and_pharmacy_percent_change_from_baseline', 'parks_percent_change_from_baseline', 'transit_stations_percent_change_from_baseline', 'workplaces_percent_change_from_baseline', 'residential_percent_change_from_baseline']

# Rename columns with a prefix
for column in mobility_columns:
    merged_date_df.rename(columns={column: f'mobility_{column}'}, inplace=True)
```

```python
print(len(merged_date_df))
```

```python
print(merged_date_df.isnull().sum())
```

```python
#Remove null values from merged data frame
merged_date_df = merged_date_df.dropna()
```

```python
print(merged_date_df.isnull().sum())
print(len(merged_date_df))
```
